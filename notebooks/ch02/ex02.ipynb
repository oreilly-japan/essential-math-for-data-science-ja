# 2章 確率

例2-1 Pythonでベイズの定理を使った例

```python
p_coffee_drinker = 0.65
p_cancer = 0.005
p_coffee_drinker_given_cancer = 0.85

p_cancer_given_coffee_drinker = p_coffee_drinker_given_cancer * \
    p_cancer / p_coffee_drinker

p_cancer_given_coffee_drinker
```

例2-2 SciPyを使った二項分布の例

```python
from scipy.stats import binom

n = 10
p = 0.9

for k in range(n + 1):
    probability = binom.pmf(k, n, p)
    print("{0} - {1}".format(k, probability))
```

例2-3 SciPyを使ったベータ分布の例

```python
from scipy.stats import beta

a = 8
b = 2

p = beta.cdf(0.90, a, b)
print(p)
```

例2-4 ベータ分布で求めた面積を引き算する例

```python
from scipy.stats import beta

a = 8
b = 2

p = 1.0 - beta.cdf(0.90, a, b)
p
```

例2-5 試行回数を増やしたベータ分布の例

```python
from scipy.stats import beta

a = 30
b = 6

p = 1.0 - beta.cdf(0.90, a, b)
print(p)
```

例2-6 SciPyでベータ分布の中央部の面積を求める例

```python
from scipy.stats import beta

a = 8
b = 2

p = beta.cdf(0.90, a, b) - beta.cdf(0.80, a, b)
print(p)
```
